### Report
Visualization of the KAFs shape learning process for both random and ridge regression initialied mixing coefficients. The experiment is done training a simple neural network for the *mnist* dataset. 

```python
# Essential Imports
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Keras imports
from tensorflow.keras import datasets
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
# Local imports
import kafnet
from kafnet import Kaf
```

```python
# Load Breast Cancer dataset
(X_train, y_train), (X_test, y_test) = datasets.mnist.load_data()

# Preprocessing is taken from here:
# https://github.com/keras-team/keras/blob/master/examples/mnist_cnn.py
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
    
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
# Batch iterator
train_data = tf.data.Dataset.from_tensor_slices((X_train.astype(np.float32), y_train.astype(np.int64)))
test_data = tf.data.Dataset.from_tensor_slices((X_test.astype(np.float32), y_test.astype(np.int64)))
```

```python
kaf_net = Sequential([
    Conv2D(32, (3, 3), input_shape=(28, 28, 1)),
    Kaf(20, conv=True, ridge='tanh'),
    Conv2D(32, (3, 3)),
    Kaf(20, conv=True, ridge='elu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(100),
    Kaf(20),
    Dense(10, activation='softmax'),
])
```

```python
kaf_net.summary()
```

```python
## How does the initial kafs look like? (before training)

d= [v for v in kaf_net.non_trainable_variables if "kaf/dictionary" in v.name][0][0][0][0][0] # Get the dictionary
a_tanh = [v for v in kaf_net.trainable_variables if "kaf/mix_coeffs" in v.name][0][0][0][0][0] # Get the a_i coefficients from the kaf tanh ridge initialized
a_elu = [v for v in kaf_net.trainable_variables if "kaf_1/mix_coeffs" in v.name][0][0][0][0][0] # Get the a_i coefficients from the kaf elu ridge initialized
a_kaf = [v for v in kaf_net.trainable_variables if "kaf_2/mix_coeffs" in v.name][0][0][0] # Get the a_i coefficients from the kaf randomly initialized

d
```

```python
# Plot the initial kafs
kb = 1/(6*(np.power(kafnet.dictionaryGen(20)[0],2)))
kaf_tan = []
kaf_elu = []
kaf_rand = []

for d_i in d:
   kaf_tan.append(kafnet.kafActivation(d_i, a_tanh, d, kb ))
   kaf_elu.append(kafnet.kafActivation(d_i, a_elu, d, kb ))
   kaf_rand.append(kafnet.kafActivation(d_i, a_kaf, d, kb ))


plt.plot(d, kaf_tan, 'r')
plt.plot(d, kaf_elu, 'g')
plt.plot(d, kaf_rand, 'b')
```

```python
# Set optimizer loss and metrics
from tensorflow.keras import losses, optimizers, metrics

opt = optimizers.Adam()
loss = losses.SparseCategoricalCrossentropy()

kaf_net.compile(optimizer=opt, loss=loss)
```

```python
# Now train the model
history = kaf_net.fit(train_data.shuffle(1000).batch(64), epochs=4)
```

```python
# Is our model accurate i.e meaningful?
acc = tf.metrics.Accuracy()
for xb, yb in test_data.batch(32):
    acc(yb, tf.argmax(kaf_net(xb), axis=1))
print('Test accuracy after epoch {} is: '+ str(acc.result()))
```

```python
# How does the kaf look like after training?
# Get learned mixing coefficients for each kaf
a_tanh_final = [v for v in kaf_net.trainable_variables if "kaf/mix_coeffs" in v.name][0][0][0][0][0]
a_elu_final= [v for v in kaf_net.trainable_variables if "kaf_1/mix_coeffs" in v.name][0][0][0][0][0] 
a_kaf_final= [v for v in kaf_net.trainable_variables if "kaf_2/mix_coeffs" in v.name][0][0][0]

kaf_tan_final = []
kaf_elu_final = []
kaf_rand_final = []

# Compute learned kafs with dictionary as activation
for d_i in d:
   kaf_tan_final.append(kafnet.kafActivation(d_i, a_tanh_final, d, kb ))
   kaf_elu_final.append(kafnet.kafActivation(d_i, a_elu_final, d, kb ))
   kaf_rand_final.append(kafnet.kafActivation(d_i, a_kaf_final, d, kb ))

# Plot kaf tanh
fig = plt.figure()
plt_tanh = fig.add_subplot()
plt_tanh.set_title('kaf tanh evolution')
plt_tanh.set_xlabel('Dictionary')
plt_tanh.plot(d, kaf_tan, 'r', label='before')
plt_tanh.plot(d, kaf_tan_final, 'r--', label='after')
plt_tanh.legend()
plt.show()

# Plot kaf elu
fig = plt.figure()
plt_elu = fig.add_subplot()
plt_elu.set_title('kaf elu evolution')
plt_elu.set_xlabel('Dictionary')
plt_elu.plot(d, kaf_elu, 'g', label='before')
plt_elu.plot(d, kaf_elu_final, 'g--', label='after')
plt_elu.legend()
plt.show()

# Plot kaf rand
fig = plt.figure()
plt_rand = fig.add_subplot()
plt_rand.set_title('kaf rand evolution')
plt_rand.set_xlabel('Dictionary')
plt_rand.plot(d, kaf_rand, 'b', label='before')
plt_rand.plot(d, kaf_rand_final, 'b--', label='after')
plt_rand.legend()
plt.show()

# Plot resulting shapes
fig = plt.figure()
plt_whole = fig.add_subplot()
plt_whole.set_title('final kafs')
plt_whole.set_xlabel('Dictionary')
plt_whole.plot(d, kaf_elu_final, 'g--', label='after')
plt_whole.plot(d, kaf_rand_final, 'b--', label='after')
plt_whole.plot(d, kaf_tan_final, 'r--', label='after')
plt.show()
```
